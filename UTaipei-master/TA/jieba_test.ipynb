# jieba

```python
import jieba
```

```python
# 看看 jieba 怎麼斷詞的
jieba.cut('今天天氣真好')
```

```python
# jieba.cut() 會回傳一個 generator，我們可以用 for 迭代看看
cut = jieba.cut('今天天氣真好')
for i in cut:
    print(i)
```

```python
# 如果想要直接回傳一個 list ，可以用 jieba.lcut
lcut = jieba.lcut('今天天氣真好')
print(lcut)
```

# 把真正的 txt 檔丟到 jieba 

```python
import os
```

```python
texts = os.listdir('./DATA')
texts
```

```python
corpus = []
for text in texts:
    with open('./DATA/{}'.format(text), 'r') as file:
        corpus.append(file.read())
```

```python
corpus_cut = [jieba.lcut(t) for t in corpus]
# 這樣就有斷完詞後的 list 了
corpus_cut
```

```python
# 斷詞後的 list 變成用空格把每個詞分開的形式
corpus = [' '.join(lcut) for lcut in corpus_cut]
# 然後就可以接到 apriori 的過程了
```
